```python
def contar_digitos(numero):
    numero = abs(numero)
    return len(str(numero))


n = int(input("Digite um número inteiro: "))
print("Quantidade de dígitos:", contar_digitos(n))
```

```python
def reverter_numero(numero):
    sinal = -1 if numero < 0 else 1

    reverso = int(str(numero)[::-1])
    return sinal * reverso


n = int(input("Digite um número inteiro: "))
print("Reverso:", reverter_numero(n))
```

```python
def data_por_extenso(data):
    meses = {
        1: "janeiro", 2: "fevereiro", 3: "março",
        4: "abril", 5: "maio", 6: "junho",
        7: "julho", 8: "agosto", 9: "setembro",
        10: "outubro", 11: "novembro", 12: "dezembro"
    }

    try:
        dia, mes, ano = map(int, data.split("/"))

        if mes not in meses:
            return None

        if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
            dias_mes = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        else:
            dias_mes = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

        if dia < 1 or dia > dias_mes[mes - 1]:
            return None

        return f"{dia} de {meses[mes]} de {ano}"

    except:
        return None


data = input("Digite a data (DD/MM/AAAA): ")
resultado = data_por_extenso(data)

if resultado:
    print(resultado)
else:
    print("Data inválida")
```

```python
def desenhar_moldura(linhas=1, colunas=1):
    linhas = max(1, min(linhas, 20))
    colunas = max(1, min(colunas, 20))

    print("+" + "-" * colunas + "+")

    for _ in range(linhas):
        print("|" + " " * colunas + "|")

    print("+" + "-" * colunas + "+")


desenhar_moldura(4, 10)
```

```python
def celsius_para_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_para_celsius(f):
    return (f - 32) * 5/9

def menu():
    print("=== Conversor de Temperatura ===")
    print("1 - Celsius para Fahrenheit")
    print("2 - Fahrenheit para Celsius")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        c = float(input("Digite a temperatura em Celsius: "))
        print("Resultado:", celsius_para_fahrenheit(c), "°F")

    elif opcao == "2":
        f = float(input("Digite a temperatura em Fahrenheit: "))
        print("Resultado:", fahrenheit_para_celsius(f), "°C")

    else:
        print("Opção inválida!")

menu()
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

def exibir_fibonacci(n, i=0):
    if i < n:
        print(fibonacci(i), end=" ")
        exibir_fibonacci(n, i + 1)


n = int(input("Digite um número inteiro positivo: "))

if n > 0:
    exibir_fibonacci(n)
else:
    print("Digite um número válido!")
```

```python
def eh_perfeito(numero):
    soma = 0

    for i in range(1, numero):
        if numero % i == 0:
            soma += i

    return soma == numero


def listar_perfeitos(n):
    for num in range(1, n + 1):
        if eh_perfeito(num):
            print(num)


n = int(input("Digite um número inteiro: "))
print("Números perfeitos até", n, ":")

listar_perfeitos(n)
```

```python
def converter_hora(hora, minuto):
    if hora == 0:
        return 12, minuto, 'A'
    elif 1 <= hora < 12:
        return hora, minuto, 'A'
    elif hora == 12:
        return 12, minuto, 'P'
    else:
        return hora - 12, minuto, 'P'


def exibir_hora(hora, minuto, periodo):
    periodo_extenso = "A.M." if periodo == 'A' else "P.M."
    print(f"{hora}:{minuto:02d} {periodo_extenso}")


while True:
    hora = int(input("Digite a hora (0-23): "))
    minuto = int(input("Digite os minutos (0-59): "))

    h, m, p = converter_hora(hora, minuto)
    exibir_hora(h, m, p)

    repetir = input("Deseja converter outra hora? (s/n): ").lower()
    if repetir != 's':
        break
```

```python
def valor_pagamento(valor, dias_atraso):
    if dias_atraso == 0:
        return valor
    else:
        multa = valor * 0.03
        juros = valor * 0.001 * dias_atraso
        return valor + multa + juros


def main():
    total = 0
    quantidade = 0

    while True:
        valor = float(input("Valor da prestação (0 para encerrar): "))

        if valor == 0:
            break

        dias = int(input("Dias em atraso: "))

        valor_final = valor_pagamento(valor, dias)

        print(f"Valor a pagar: R$ {valor_final:.2f}\n")

        total += valor_final
        quantidade += 1

    print("\n=== Relatório do dia ===")
    print(f"Quantidade de prestações: {quantidade}")
    print(f"Total arrecadado: R$ {total:.2f}")


main()
```

```python
pedidos = []

def cadastrar_pedido():
    nome = input("Nome do cliente: ")
    item = input("Item pedido: ")
    quantidade = int(input("Quantidade: "))
    preco = float(input("Preço unitário: "))

    pedido = {
        "nome": nome,
        "item": item,
        "quantidade": quantidade,
        "preco": preco
    }

    pedidos.append(pedido)
    print("Pedido cadastrado com sucesso!\n")


def listar_pedidos():
    if not pedidos:
        print("Nenhum pedido cadastrado.\n")
        return

    for i, p in enumerate(pedidos):
        total = p["quantidade"] * p["preco"]
        print(f"{i} - Cliente: {p['nome']}, Item: {p['item']}, "
              f"Qtd: {p['quantidade']}, Preço: R$ {p['preco']:.2f}, "
              f"Total: R$ {total:.2f}")
    print()


def atualizar_pedido():
    listar_pedidos()

    if not pedidos:
        return

    i = int(input("Digite o índice do pedido a atualizar: "))

    if 0 <= i < len(pedidos):
        print("Deixe em branco para não alterar.")

        nome = input("Novo nome: ")
        item = input("Novo item: ")
        quantidade = input("Nova quantidade: ")
        preco = input("Novo preço: ")

        if nome:
            pedidos[i]["nome"] = nome
        if item:
            pedidos[i]["item"] = item
        if quantidade:
            pedidos[i]["quantidade"] = int(quantidade)
        if preco:
            pedidos[i]["preco"] = float(preco)

        print("Pedido atualizado!\n")
    else:
        print("Índice inválido!\n")


def remover_pedido():
    listar_pedidos()

    if not pedidos:
        return

    i = int(input("Digite o índice do pedido a remover: "))

    if 0 <= i < len(pedidos):
        pedidos.pop(i)
        print("Pedido removido!\n")
    else:
        print("Índice inválido!\n")


def menu():
    while True:
        print("=== Sistema de Pedidos ===")
        print("1 - Cadastrar pedido")
        print("2 - Listar pedidos")
        print("3 - Atualizar pedido")
        print("4 - Remover pedido")
        print("5 - Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            cadastrar_pedido()
        elif opcao == "2":
            listar_pedidos()
        elif opcao == "3":
            atualizar_pedido()
        elif opcao == "4":
            remover_pedido()
        elif opcao == "5":
            print("Encerrando sistema...")
            break
        else:
            print("Opção inválida!\n")


menu()
```
